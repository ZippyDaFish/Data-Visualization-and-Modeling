```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('titanic_passengers.csv', encoding='ISO-8859-1')
```

### Character Profile 1: Most Likely to Survive
- Sex: Female (women had much higher survival rates).
- Age: Child or young adult (children prioritized, and younger adults more resilient).
- Class: 1st class (wealthier passengers had better access to lifeboats).
- Fare: Higher fare served as a proxy for upper class.
- Travel Companions: With family since being with parents/children often increased survival chances.
- Embarked: Cherbourg had slightly higher survival rates since many wealthy passengers boarded here.

Character Concept: A young 1st class woman traveling with her parents, who boarded at Cherbourg. She’s wealthy, well-dressed, and quickly escorted to the lifeboats.

### Character Profile 2: Least Likely to Survive
- Sex: Male (men were less prioritized).
- Age: Adult (not a child, but not elderly enough to get sympathy).
- Class: 3rd class (poor access to lifeboats, language barriers).
- Fare: Low fare indicated the poorest passengers.
- Travel Companions: Alone (no one to advocate for them).
- Embarked: Queenstown or Southampton, where survival rates were lower.

Character Concept: A young man in his 20s, traveling alone in 3rd class from Queenstown. He had little chance of reaching the deck in time.

```python
plt.figure(figsize=(6,4))
df.groupby("Sex")["Survived"].mean().plot(kind="bar")
plt.title("Survival Rate by Sex")
plt.ylabel("Survival Rate")
plt.xlabel("Sex")
plt.show()
```

```python
plt.figure(figsize=(8,5))
df["AgeGroup"] = pd.cut(df["Age"], bins=[0,12,18,30,50,80], labels=["Child","Teen","Young Adult","Adult","Elderly"])
df.groupby("AgeGroup")["Survived"].mean().plot(kind="bar")
plt.title("Survival Rate by Age Group")
plt.ylabel("Survival Rate")
plt.xlabel("Age Group")
plt.show()
```

```python
plt.figure(figsize=(6,4))
df.groupby("Pclass")["Survived"].mean().plot(kind="bar")
plt.title("Survival Rate by Passenger Class")
plt.ylabel("Survival Rate")
plt.xlabel("Class (1=Upper, 2=Middle, 3=Lower)")
plt.show()
```

```python
plt.figure(figsize=(8,5))
df["FareGroup"] = pd.qcut(df["Fare"], 4, labels=["Low","Medium","High","Very High"])
df.groupby("FareGroup")["Survived"].mean().plot(kind="bar")
plt.title("Survival Rate by Fare Group")
plt.ylabel("Survival Rate")
plt.xlabel("Fare Group")
plt.show()
```

```python
plt.figure(figsize=(8,5))
df["TravelCompanions"] = df["SibSp"] + df["Parch"]
df.groupby("TravelCompanions")["Survived"].mean().plot(kind="bar")
plt.title("Survival Rate by Number of Travel Companions")
plt.ylabel("Survival Rate")
plt.xlabel("Number of Companions")
plt.show()
```

```python
plt.figure(figsize=(6,4))
df.groupby("Embarked")["Survived"].mean().plot(kind="bar")
plt.title("Survival Rate by Port of Embarkation")
plt.ylabel("Survival Rate")
plt.xlabel("Port (C=Cherbourg, Q=Queenstown, S=Southampton)")
plt.show()
```
